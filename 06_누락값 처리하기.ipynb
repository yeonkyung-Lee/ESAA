
## 누락값의 개수 구하기

#### count( ) : 누락값이 아닌 경우 갯수 세기

```python
print(ebola.count())
ebola.shape[0]
```

### isnull(  ): nan값 검사하기 T/F, count_nonzero(  ): 0이 아닌값 세기 

```python
import numpy as np 

print(np.count_nonzero(ebola.isnull()))  #안에 있는 모든 누락값의 개수
```

```python
#Case_Guinea에 대해서 nan값 검사해서 숫자 세기
np.count_nonzero(ebola(['Cases_Guinea'],isnull()))
```

### value_counts(  ): 지정한 열의 값 빈도수 
#### Cases_Guinea 열의 결측값 갯수 세기

```python
print(ebola.Cases_Guinea.value_counts(dropna=False).head())
```


## 누락값 처리하기 ― 변경, 삭제

### 1. 누락값 변경하기

### fillna(  ) : 괄호안의 값으로 누락값 대체

```python
print(ebola.fillna(0).iloc[0:10, 0:5])  #결측값을 0으로 채움
```

### fillna(method='ffill') : 누락값이 나타나기 전의 값으로 변경

```python
print(ebola.fillna(method='ffill').iloc[0:10, 0:5])
```

### fillna(method='bfill') : 누락값이 나타난 이후의 첫 번째 값으로 변경

```python
print(ebola.fillna(method='bfill').iloc[0:10, 0:5])
```

### interpolate(  ): 누락값 양쪽의 중간값으로 처리

```python
print(ebola.interpolate().iloc[0:10, 0:5])
```

### 평균값으로 채우기

```python
#에볼라 데이터의 결측값을 평균값으로 채우기
ebola.fillna(ebola.mean())
```

### 5. dropna(   ) : 누락값 삭제하기

```python
print(ebola.shape)
```

```python
ebola_dropna = ebola.dropna() 
print(ebola_dropna.shape)  #결측값이 하나라도 있으면 다 삭제됨
```

```python
print(ebola_dropna)
```

#### how='all' 모든 행의 값이 NaN인 경우 삭제

```python
data= pd.DataFrame([[1.,6.5,3.],[1.,NAN,NAN],[NAN,NAN,NAN],[NAN,6.5,3.]])
data
```

```python
data.dropna(how='all')   #전부 결측치였던 2번 행만 삭제됨.
```

#### thresh = n 행에 값이 n개 이상인 것만 골라내기

```python
data.dropna(thresh=2)
```

```python
#에볼라 데이터에서 결측값이 5개 이상인 것 삭제하기
ebola.dropna(thresh=5).shape
```


## 누락값이 포함된 데이터 계산하기

```python
ebola['Cases_multiple'] = ebola['Cases_Guinea'] + ebola['Cases_Liberia'] + ebola['Cases_SierraLeone']
```

```python
ebola_subset = ebola.loc[:, ['Cases_Guinea', 'Cases_Liberia', 'Cases_SierraLeone', 'Cases_multiple']] 
print(ebola_subset.head(n=10))
```

```python
print(ebola.Cases_Guinea.sum(skipna = True))
```

```python
print(ebola.Cases_Guinea.sum(skipna = False))
```

```python
#앞서 ebola data에 대한 더하기를 행기준으로 수행하고 'sum'이라는 컬럼에 저장해 새로운 데이터 프레임 만들기
new_data=ebola.loc[:,['Cases_Guinea','Cases_Liberia','Cases_SierraLeone']]
new_data['Cases_multiple2']=new_data.sum(axis=1,skipna=True)  #skipna=True로 인해 결측치는 더하지 않음.
new_data
```
