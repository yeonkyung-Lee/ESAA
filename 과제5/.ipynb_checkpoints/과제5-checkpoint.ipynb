```python
import pandas as pd
GB=pd.read_csv('C:/Users/SAMSUNG/Desktop/GBvideos.csv')
US=pd.read_csv('C:/Users/SAMSUNG/Desktop/USvideos.csv')
```

```python
GB.head()  #영국데이터 이용
```

```python
#category_id별 likes 평균 구하기
avg_likes_by_category_id=GB.groupby('category_id').likes.mean()
print(avg_likes_by_category_id)
```

```python
#category id별 likes 최대값 구하기
max_likes_by_category_id=GB.groupby('category_id').likes.max()
print(max_likes_by_category_id)
```

```python
#category id별 likes 최소값 구하기
min_likes_by_category_id=GB.groupby('category_id').likes.min()
print(min_likes_by_category_id)
```

```python
#category_id별 comment_count 평균 구하기
avg_comment_count_by_category_id=GB.groupby('category_id').comment_count.mean()
print(avg_comment_count_by_category_id)
```

```python
#category_id별 comment_count 최대값 구하기
max_comment_count_by_category_id=GB.groupby('category_id').comment_count.max()
print(max_comment_count_by_category_id)
```

```python
#category_id별 comment_count 최소값 구하기
min_comment_count_by_category_id=GB.groupby('category_id').comment_count.min()
print(min_comment_count_by_category_id)
```
